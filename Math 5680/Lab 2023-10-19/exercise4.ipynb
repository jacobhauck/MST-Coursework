```python
import imageio
import matplotlib.pyplot as plt
import numpy as np

image = imageio.v3.imread('swim.jpg')
plt.imshow(image)
plt.show()
```

```python
from sklearn.cluster import KMeans

colors = image.reshape(-1, 3).astype(float) / 255.
kmeans = KMeans(5).fit(colors)

image2 = kmeans.cluster_centers_[kmeans.labels_, :].reshape(image.shape)
image2 = (image2 * 255).astype(np.uint8)

plt.imshow(image2)
plt.show()
imageio.v3.imwrite('exercise4.jpg', image2)
```
