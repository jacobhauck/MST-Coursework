```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('ruspini_data.txt')
print(f'Loaded data shape: {data.shape}')
print(f'First five rows')
print(data[:5])

data2 = (data - np.mean(data, axis=0)) / np.std(data, axis=0)

_, ax = plt.subplots()
ax.scatter(data2[:, 0], data2[:, 1])
plt.show()
```

```python
from sklearn.cluster import KMeans

E = []
for k in range(1, 11):
    E.append(KMeans(k).fit(data2).inertia_)
    print(f'k = {k}, Energy = {E[k-1]}')

_, ax = plt.subplots()
ax.scatter(range(1, 11), E)
plt.show()
```

```python
# Choose k = 4!

good_kmeans = KMeans(4).fit(data2)

_, ax = plt.subplots()
c = np.array(['red', 'blue', 'green', 'purple'])[good_kmeans.labels_]
ax.scatter(data2[:, 0], data2[:, 1], c=c)

ax.scatter(
    good_kmeans.cluster_centers_[:, 0],
    good_kmeans.cluster_centers_[:, 1],
    c='black', marker='*'
)
plt.show()
plt.savefig('exercise3.jpg')
```
