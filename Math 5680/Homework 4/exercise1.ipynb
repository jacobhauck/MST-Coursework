```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('hw_data.txt')
print(f'Loaded array shape: {data.shape}')
print()

print('First five rows of data')
print(data[:5])
print()

height_in = data[:350, 1]
weight_lbs = data[:350, 2]


def print_summary(title, x):
    print(title)
    print(f'Min = {np.min(x):.02f}, Max = {np.max(x):.02f}')
    print(f'Mean = {np.mean(x):.02f}, Var = {np.var(x)}, Std = {np.std(x)}')
    print()
    
    
print_summary('Height summary (in)', height_in)
print_summary('Weight summary (lbs)', weight_lbs)

height_std = (height_in - np.mean(height_in)) / np.std(height_in)
weight_std = (weight_lbs - np.mean(weight_lbs)) / np.std(weight_lbs)

_, ax = plt.subplots()
ax.scatter(height_std, weight_std)

t = np.linspace(0, 2 * np.pi, 51)
x, y = np.cos(t), np.sin(t)

for radius in (1, 2, 3):
    ax.plot(radius * x, radius * y, 'r', linewidth=2)

ax.set_aspect(1.0)
    
plt.show()
plt.savefig('exercise1.jpg')

# We can also standardize using matrix square root of
top_350 = data[:350]
cov = np.cov(top_350[:, 0], top_350[:, 1])
u, d, vt = np.linalg.svd(cov)
cov_sqrt_inv = u @ d ** (-.5) @ vt
data2 = (cov_sqrt_inv @ (top_350 - np.mean(top_350, axis=0, keepdims=True)).T
```
